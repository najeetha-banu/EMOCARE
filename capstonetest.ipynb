```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from textblob import TextBlob
import nltk
from sklearn.metrics import accuracy_score

nltk.download('punkt')

# Load the trained emotion recognition model
model_path = '/content/drive/MyDrive/emotion_recognition_model.h5'  # Update to your model path
emotion_model = load_model(model_path)

# Emotion labels for facial emotion recognition
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# Preprocess the facial image for emotion recognition
def preprocess_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized_img = cv2.resize(gray, (48, 48))
    normalized_img = resized_img / 255.0
    normalized_img = np.stack([normalized_img] * 3, axis=-1)
    input_img = np.expand_dims(normalized_img, axis=0)
    return input_img

# Perform sentiment analysis on extracted text
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Combine predictions from both image and text
def final_prediction(image_prediction, text_prediction):
    if image_prediction == 'Happy' or text_prediction == 'Positive':
        return 'Positive'
    elif image_prediction in ['Angry', 'Disgust', 'Fear', 'Sad'] or text_prediction == 'Negative':
        return 'Negative'
    else:
        return 'Neutral'

# Function to evaluate accuracy on a test dataset
def evaluate_model(test_images, true_labels):
    predictions = []

    for img_path in test_images:
        img = cv2.imread(img_path)
        input_image = preprocess_image(img)
        preds = emotion_model.predict(input_image)
        predicted_emotion = emotion_labels[np.argmax(preds)]
        predictions.append(predicted_emotion)

    accuracy = accuracy_score(true_labels, predictions)
    return accuracy

# Read the uploaded image for facial emotion detection
img_path = '/content/drive/MyDrive/1.jpg'  # Path to the uploaded image
img = cv2.imread(img_path)
input_image = preprocess_image(img)

# Predict the emotion from the image
predictions = emotion_model.predict(input_image)
predicted_emotion = emotion_labels[np.argmax(predictions)]

# Read the text from the uploaded file
with open('/content/drive/MyDrive/3.txt', 'r', encoding='utf-8') as f:
    extracted_text = f.read()

# Analyze sentiment from the text
text_sentiment = analyze_sentiment(extracted_text)

# Combine both predictions
final_result = final_prediction(predicted_emotion, text_sentiment)

# Output the results
print(f'Predicted Emotion from Image: {predicted_emotion}')
print(f'Extracted Text from File: {extracted_text}')
print(f'Sentiment from Text: {text_sentiment}')
print(f'Final Prediction (Combined): {final_result}')

```
